```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import pandas as pd

# 1. Loading the Iris dataset
# Iris has 4 features: Sepal Length, Sepal Width, Petal Length, Petal Width
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# 2. Implementing PCA
# We reduce the 4D data down to 2D so we can plot it on a graph
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# 3. Formatted Output for Lab Record
print("Output:")
print(f"\nOriginal Shape: {X.shape}")
print(f"Reduced Shape: {X_pca.shape}")

# Explained Variance shows how much information was kept
print(f"\nExplained Variance Ratio: {pca.explained_variance_ratio_}")
print(f"Total Variance Retained: {sum(pca.explained_variance_ratio_):.2f}")

# 4. Visualizing the Principal Components
plt.figure(figsize=(8, 6))
colors = ['navy', 'turquoise', 'darkorange']

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], color=color, alpha=.8, lw=2,
                label=target_name)

plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

print("\nResult: Dimensionality Reduction using PCA Complete!")
```
